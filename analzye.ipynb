```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_sd = pd.read_csv('data/data-sd.txt', sep='\t', header=None)
df_sd = df_sd.loc[:, [1, 3, 5]]
df_sd = df_sd.rename(columns={1: 'provinsi', 3: 'jumlah_siswa', 5: 'putus_sekolah'})

# Add ratio_putus_sekolah
df_sd['ratio_putus_sekolah_sd'] = df_sd['putus_sekolah'] / df_sd['jumlah_siswa']

df_sd
```

```python
df_smp = pd.read_csv('data/data-smp.txt', sep='\t', header=None)
df_smp = df_smp.loc[:, [1, 3, 5]]
df_smp = df_smp.rename(columns={1: 'provinsi', 3: 'jumlah_siswa', 5: 'putus_sekolah'})

# Add ratio_putus_sekolah
df_smp['ratio_putus_sekolah_smp'] = df_smp['putus_sekolah'] / df_smp['jumlah_siswa']

df_smp
```

```python
df_sma = pd.read_csv('data/data-sma.txt', sep='\t', header=None)
df_sma = df_sma.loc[:, [1, 3, 5]]
df_sma = df_sma.rename(columns={1: 'provinsi', 3: 'jumlah_siswa', 5: 'putus_sekolah'})

# Add ratio_putus_sekolah
df_sma['ratio_putus_sekolah_sma'] = df_sma['putus_sekolah'] / df_sma['jumlah_siswa']

df_sma
```

```python
# Merge df_sd and df_smp
merged_df = pd.merge(df_sd[['provinsi', 'ratio_putus_sekolah_sd']], df_smp[['provinsi', 'ratio_putus_sekolah_smp']], on='provinsi', how='inner')

# Merge merged_df with df_sma
merged_df = pd.merge(merged_df, df_sma[['provinsi', 'ratio_putus_sekolah_sma']], on='provinsi', how='inner')

merged_df
```

```python
print("Rata-rata ratio putus sekolah sd : ", end='')
print(merged_df['ratio_putus_sekolah_sd'].mean())

print("Rata-rata ratio putus sekolah smp : ", end='')
print(merged_df['ratio_putus_sekolah_smp'].mean())

print("Rata-rata ratio putus sekolah sma : ", end='')
print(merged_df['ratio_putus_sekolah_sma'].mean())

print("Standar deviasi ratio putus sekolah sd : ", end='')
print(merged_df['ratio_putus_sekolah_sd'].std())

print("Standar deviasi ratio putus sekolah smp : ", end='')
print(merged_df['ratio_putus_sekolah_smp'].std())

print("Standar deviasi ratio putus sekolah sma : ", end='')
print(merged_df['ratio_putus_sekolah_sma'].std())
```

```python
# Data
provinsi = merged_df['provinsi']
ratio_putus_sekolah_sd = merged_df['ratio_putus_sekolah_sd']
ratio_putus_sekolah_smp = merged_df['ratio_putus_sekolah_smp']
ratio_putus_sekolah_sma = merged_df['ratio_putus_sekolah_sma']

# Create a DataFrame for the chart
data = pd.DataFrame({
    'Provinsi': provinsi,
    'SD': ratio_putus_sekolah_sd,
    'SMP': ratio_putus_sekolah_smp,
    'SMA': ratio_putus_sekolah_sma
})

data_long = pd.melt(data, id_vars='Provinsi', var_name='Education Level', value_name='Ratio')
plt.figure(figsize=(10, 0.5 * len(provinsi)))
sns.barplot(data=data_long, y='Provinsi', x='Ratio', hue='Education Level', alpha=0.7)
plt.xlabel('Rasio')
plt.title('Rasio Putus Sekolah tingkat SD, SMP, dan SMA di tiap provinsi di Indonesia')
plt.legend()

plt.show()
```

```python
df_kemiskinan = pd.read_excel("data/rate-kemiskinan.xlsx")

df_kemiskinan_sorted = df_kemiskinan.sort_values('Persentase', ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x='Persentase', y='Provinsi', data=df_kemiskinan_sorted, color='skyblue')
plt.xlabel('Persentase (%)')
plt.ylabel('Provinsi')
plt.title('Persentase Kemiskinan per Provinsi')

plt.show()
```

```python
print("Rata-rata persentase kemiskinan : ", end='')
print(df_kemiskinan['Persentase'].mean())

print("Standar deviasi persentase kemiskinan : ", end='')
print(df_kemiskinan['Persentase'].std())
```
